### Dataset 2: Numbeo - Cost of Living

The price of Airbnb rentals may be affected by the cost of living in different cities and countries. For example, a one bedroom apartment in New York will cost considerably more in comparison to the cost of the same sized apartment in Hanoi, Vietnam. Therefore the infulence of the cost of living should be considered when analysing Airbnb rental prices.

Numbeo maintains a global database with a plethora of information pertaining to quality of life indicators, which include but are not limited to the cost of living, housing prices, healthcare quality, and crime rates. One such dataset available on Kaggle, compiled by Numbeo, summarises the prices of goods and services into indices; Cost of Living, Rent, Groceries, Restaurant Price and Local Purchasing Power. These provide an overview of the cost of living.

The indices measure the difference relative to New York City. The value of each of the indices for New Your City are 100%. If, for instance, the rent index for a certain city is 80, this means that, on average, the rent in that city is 20% less expensive than in New York City.

The Cost of Living data on Numbeo is categorised into various segments covering a wide range of everyday expenses, such as restaurants, markets, transportation, utilities, sports and leisure, childcare, clothing and shoes, monthly rent, buying property and salaries. These categories encompass most of the day-to-day expenditures for a significant portion of the population, thus providing a useful indicator of the cost of living.

The restaurant category contains details about the expenses associated with eating out, covering the typical cost of dining at a low-cost restaurant and a three-course meal for two at a mid-level establishment, as well as the costs of beverages served in restaurants and cafes. The markets section pertains to the typical grocery items that are readily available at supermarkets across the globe. This includes fresh produce such as milk (1L), a loaf of bread (500g), as well as fruits and vegetables like apples, bananas, onions, and potatoes sold per kilo.

The transportation section comprises information regarding the expenses of local transport,
featuring the cost of local transport and taxi fares, and fuel(1L). Utilities include household bills and sports and leisure include gym memberships and the cost of cinema tickets.

The measurement of rental costs involves assessing the prices of apartments with one and three bedrooms located both within and outside the city centre. In order to evaluate the cost of purchasing real estate, the price per square meter is taken into consideration for both city centre and non-city centre locations. Additionally, financial data is analysed by taking into account the average monthly salary (after taxes) and the typical annual mortgage interest rate (as a percentage) for a 20-year fixed-rate mortgage.

The Cost of Living Index is a relative indicator of the costs of groceries, restaurants, transportation and utilities. This index does not factor in rent or mortgage payments. The Rent index is a prediction of the cost of renting an apartment relative to the cost in New York.

The Groceries Index is an estimation of typical grocery items and the Restaurant Index is a comparison of the prices of meals and beverages in restaurants and bars.  

The Cost of Living Plus Rent Index factors in the cost of renting/ paying a mortgage in a certain city, relative to the costs in New Your City.

The Local Purchasing Power measures the purchasing power in buying goods and services in a city for their average net salary. For example, if purchasing power is 60 for some city, then the individuals who earn the average salary for that city can afford to purchase, on average, 40% fewer goods and services than New York City residents who earn an average salary for New York City.

Further information on thos data can be found [here](https://www.kaggle.com/datasets/ankanhore545/cost-of-living-index-2022?select=Cost_of_Living_Index_2022).

### Dataset 3: Numbeo - Crime Index
In addition to location and day of the week, another important factor that may impact the rental prices of Airbnb properties is the safety of the city they are located. A city's safety level can affect the number of visitors it attracts, which in turn can impact the demand for rentals and drive prices higher or lower. To measure the safety of a city, we can use the Crime Index, which provides an estimate of the extent of crime in a given location.

The Crime Index is typically calculated on a scale of 0 to 100 (float64 data type), with lower scores indicating a safer city. For example, a score of 20 or below is considered very low (i.e. very safe), while a score above 80 is considered very high (i.e. very unsafe). In between, scores of 20-40 is viewed as low, 40-60 is considered moderate, and 60-80 is deemed high.

To explore the relationship between safety and Airbnb rental prices, we can incorporate the Crime Index into our dataset. One source of this data is Numbeo, which provides a comprehensive dataset containing the Crime and Safety Indexes of 453 cities across the globe. Specifically, we can use the 2022 Crime Index dataset found on Kaggle, which includes scores for a wide range of cities.

It is worth noting that the data also cotains a Safety Index, which measures the opposite of the Crime Index and reflects a city's overall safety level. The Safety Index is also typically measured on a scale of 0 to 100, with higher scores indicating a safer city. Because the sum of the Crime and Safety Indexes is always 100, we can choose to use either one in our analysis.

Further information can be found [here](https://www.numbeo.com/crime/rankings.jsp).

#### _Brief Explanation of following code cell_
Adding a column to each of the datasets called `day` that is populated with `weekday` or `weekend` depending on what day the data is for. This will ensure we can differentiate between weekday and weekend data when consolidating the dataset.

```python
# Adding a column named 'day' to each dataframe and populating with 'weekday' or 'weekend'
weekdays = [amsterdam_weekday, athens_weekday, barcelona_weekday, berlin_weekday, budapest_weekday,
            lisbon_weekday, london_weekday, paris_weekday, rome_weekday, vienna_weekday]
weekends = [amsterdam_weekend, athens_weekend, barcelona_weekend, berlin_weekend, budapest_weekend,
            lisbon_weekend, london_weekend, paris_weekend, rome_weekend, vienna_weekend]
for df in weekdays:
    df.insert(1, 'day', 'weekday')

for df in weekends:
    df.insert(1, 'day', 'weekend')
```

#### _Brief Explanation of following code cell_
Defining a function that adds a new `city` and `country` columns to the dataframe.

```python
# Function that adds new columns for country and city names to the data frame
def city_country_name_col(data_frame, country_name, city_name, position):
    data_frame.insert(position, 'country', country_name)
    data_frame.insert(position+1, 'city', city_name)
    return data_frame
```

#### _Brief Explanation of following code cell_
Joining the datasets for each city since we have added a column to differentiate between weekday and weekend data.

```python
# Joining weekday and weekend data
amsterdam = pd.concat([amsterdam_weekday, amsterdam_weekend], axis = 0)
athens = pd.concat([athens_weekday, athens_weekend], axis = 0)
barcelona = pd.concat([barcelona_weekday, barcelona_weekend], axis = 0)
berlin = pd.concat([berlin_weekday, berlin_weekend], axis = 0)
budapest = pd.concat([budapest_weekday, budapest_weekend], axis = 0)
lisbon = pd.concat([lisbon_weekday, lisbon_weekend], axis = 0)
london = pd.concat([london_weekday, london_weekend], axis = 0)
paris = pd.concat([paris_weekday, paris_weekend], axis = 0)
rome = pd.concat([rome_weekday, rome_weekend], axis = 0)
vienna = pd.concat([vienna_weekday, vienna_weekend], axis = 0)
```

#### _Brief Explanation of following code cell_
The code below defines a function that creates a summary table of average price and average log prices of listings in each city. We will call the function later in the report.

```python
cities = list(airbnb['city'].unique())
def average_price_table():
    # Creating dictionaries for the average price and average log price for each city
    cities = list(airbnb['city'].unique())
    average_price = {}
    average_log_price = {}
    for c in cities:
        average_price[c] = round(airbnb[airbnb['city'] == c]['price'].mean(),2)
        average_log_price[c] = round(airbnb[airbnb['city'] == c]['log_price'].mean(),2)

    # Converting the dictionaries into a dataframe
    df_average_price = pd.DataFrame(
        {
            'Average Price': list(average_price.values()),
            'Average Log Price' : list(average_log_price.values())
        },
        index = cities)

    return df_average_price
```

#### _Brief Explanation of following code cell_
The following code defines a function that creates a summary table of the counts of each room type in each city. We will call this function later in the report.

```python
def room_count_table():
    # Creating dictionaries for the counts for the room types in each city
    cities = list(airbnb['city'].unique())
    rooms = list(airbnb['room_type'].unique())
    room_count = {}
    for c in cities:
        room_count[c] = {}
        for r in rooms:
            room_count[c][r] = len(airbnb[(airbnb['city'] == c) & (airbnb['room_type'] == r)])

    # Converting the dictionary into a data frame
    df_room_count = pd.DataFrame.from_dict(room_count, orient='index')

    return df_room_count
```

#### _Brief Explanation of following code cell_
The following cell defines a function that creates a summary table displaying the average price and average log prices for listings with different capacities for each city. The function will be useful later in the report.

```python
def capacity_summary_table():
    # Creating dictionaries for the average price and log price for listings with different capacities
    cities = list(airbnb['city'].unique())
    capacity = sorted(list(airbnb['person_capacity'].unique()))
    capacity_price = {}
    capacity_log_price = {}
    for c in cities:
        capacity_price[c] = {}
        capacity_log_price[c] = {}
        for cap in capacity:
            capacity_price[c][cap] = round(airbnb[(airbnb['city'] == c) & (airbnb['person_capacity'] == cap)]['price'].mean(),2)
            capacity_log_price[c][cap] = round(airbnb[(airbnb['city'] == c) & (airbnb['person_capacity'] == cap)]['log_price'].mean(),2)

    # Converting the dictionaries into data frames and melting/pivoting for better presentation
    capacity_price = pd.DataFrame.from_dict(capacity_price, orient='index')
    capacity_log_price = pd.DataFrame.from_dict(capacity_log_price, orient='index')

    capacity_price = capacity_price.melt(ignore_index = False)
    capacity_log_price = capacity_log_price.melt(ignore_index = False)

    capacity_price.rename(columns = {'variable': 'capacity', 'value':'Average Price'}, inplace = True)
    log_prices = list(capacity_log_price['value'])
    capacity_price.insert(2,'Average Log Price', log_prices)

    capacity_price = capacity_price.pivot(columns = 'capacity')
    return capacity_price
```

#### _Brief Explanation of following code cell_
The following code defines a function to create a correlation heatmap, taking inputs of the data frame and the figure size. It also encodes given categorical columns in the dataframe. The function will be used later in the report.

```python
def generate_heatmap(dataframe, figsize, cat_cols):
    # Identifying categorical columns we need to encode
    obj_cols = cat_cols

    # Copying the data frame into a new variable
    df_cat = dataframe.copy()

    # Encode categorical columns with LabelEncoder
    for col in obj_cols:
        le = LabelEncoder()
        df_cat[col] = le.fit_transform(df_cat[col])


    # save correlations to variable
    corr = df_cat.corr()

    # we can create a mask to not show duplicate values
    mask = np.zeros_like(corr, dtype=np.bool)
    mask[np.triu_indices_from(mask)] = True

    # generate heatmap
    plt.figure(figsize= figsize)
    sns.heatmap(corr, annot=True, center=0, mask=mask, cmap='gnuplot')
    plt.show()
```
